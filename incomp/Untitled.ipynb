```python
# コマンド実行 ルーチン
def exec_cmd(cmd):
    from subprocess import Popen, PIPE

    p = Popen(cmd.split(' '), stdout=PIPE, stderr=PIPE)
    out, err = p.communicate()

    return [ s for s in out.split('\n') if s ]
```
